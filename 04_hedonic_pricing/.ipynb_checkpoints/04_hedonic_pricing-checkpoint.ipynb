```python
import pandas as pd

df = pd.read_csv('cars_hw.csv')
df.describe()
```

```python
df.head()
```

```python
df_clean = df.copy()
df_clean['Make_Year'] = pd.to_numeric(df_clean['Make_Year'], errors='coerce')
df_clean['Seating_Capacity'] = pd.to_numeric(df_clean['Seating_Capacity'], errors='coerce')
df_clean = df_clean.drop_duplicates()
# These prices all seem quite high, I am working under the assumption that the data is malformed and missing a decimal as that would make the numbers more realistic.
df_clean['Price'] = df_clean['Price'] / 10
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Asked Claude for help with generating graphs by providing which metrics I wanted to compare. 
sns.histplot(data=df_clean, x='Price', bins=30)
plt.title('Distribution of Car Prices')
plt.xlabel('Price')
plt.ylabel('Count')
plt.show()

sns.scatterplot(data=df_clean, x='Mileage_Run', y='Price', alpha=0.5)
plt.title('Price vs Mileage')
plt.xlabel('Mileage')
plt.ylabel('Price')
plt.show()

df_clean.groupby('Make')['Price'].mean().sort_values(ascending=False).head(10).plot(kind='bar')
plt.title('Average Price by Make')
plt.xlabel('Make')
plt.ylabel('Average Price')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
N = df_clean.shape[0]
df_clean = df_clean.sample(frac=1, random_state=100)
train_size = int(.8*N)

df_train = df_clean[0:train_size]
y_train = df_train['Price']

df_test = df_clean[train_size:]
y_test = df_test['Price']
```

```python
categories = ['Make','Body_Type','Color','Fuel_Type','Transmission','Transmission_Type']

dummies = pd.DataFrame([])
for category in categories: 
    new_dummies = pd.get_dummies(df_clean.loc[:,category], drop_first=True, dtype=int)
    dummies = pd.concat([dummies, new_dummies], axis=1, ignore_index=True) 
    
X_train_c = dummies.iloc[0:train_size,:]
X_test_c = dummies.iloc[train_size:,:]
```

```python
from sklearn import linear_model
import numpy as np

var_n = ['Make_Year','Mileage_Run','Seating_Capacity'] 
X_train_n = df_train.loc[:,var_n] 
reg_n = linear_model.LinearRegression().fit(X_train_n,y_train)
X_test_n = df_test.loc[:,var_n]
y_hat = reg_n.predict(X_test_n)

r2_numeric = reg_n.score(X_test_n,y_test)
rmse_numeric = np.sqrt( np.mean( (y_test - y_hat)**2 ))
```

```python
reg_c = linear_model.LinearRegression().fit(X_train_c,y_train) 
y_hat_c = reg_c.predict(X_test_c)

r2_categorical = reg_c.score(X_test_c,y_test)
rmse_categorical = np.sqrt( np.mean( (y_test - y_hat_c)**2 ))
```

```python
X_train_all = pd.concat([X_train_n, X_train_c],axis=1, ignore_index=True)
X_test_all = pd.concat([X_test_n, X_test_c],axis=1, ignore_index=True)

reg_all = linear_model.LinearRegression().fit(X_train_all,y_train) 
y_hat_all = reg_all.predict(X_test_all)

r2_all = reg_all.score(X_test_all,y_test)
rmse_all = np.sqrt( np.mean( (y_test - y_hat_all)**2 ))
```

```python
print(r2_numeric)
print(rmse_numeric)
print(r2_categorical)
print(rmse_categorical)
print(r2_all)
print(rmse_all)
```

Regressing off numerical data gives us 0.35, regressing off categorical gives us 0.71, and regressing off both numeric and categorical gives us 0.81. So, it appears regressing off all our data attributes in this particular case is the best overall. We can also see our RMSE goes down from 28k, to 19k, to 15k, so our variance is dropping as well.

```python
from sklearn.preprocessing import StandardScaler

X_train_n['Mileage_Run_log'] = np.log(X_train_n['Mileage_Run'] + 1)  
X_train_n['Make_Year_sq'] = X_train_n['Make_Year'] ** 2  
X_test_n['Mileage_Run_log'] = np.log(X_test_n['Mileage_Run'] + 1)
X_test_n['Make_Year_sq'] = X_test_n['Make_Year'] ** 2

scaler = StandardScaler()
X_train_n_scaled = scaler.fit_transform(X_train_n)
X_test_n_scaled = scaler.transform(X_test_n)

X_train_c_encoded = pd.get_dummies(X_train_c, drop_first=True)
X_test_c_encoded = pd.get_dummies(X_test_c, drop_first=True)

X_train_c_encoded, X_test_c_encoded = X_train_c_encoded.align(X_test_c_encoded, join='left', axis=1, fill_value=0)

X_train_all = pd.concat([pd.DataFrame(X_train_n_scaled), X_train_c_encoded.reset_index(drop=True)], axis=1)
X_test_all = pd.concat([pd.DataFrame(X_test_n_scaled), X_test_c_encoded.reset_index(drop=True)], axis=1)

reg_all = linear_model.LinearRegression().fit(X_train_all, y_train)

y_hat = reg_all.predict(X_test_all)
rmse = np.sqrt(np.mean((y_test - y_hat) ** 2))
print(reg_all.score(X_test_all,y_test))
print(rmse)
```

After doing some numerical scaling, it does appear to help our model but only slightly, increasing our RSQ to a slightly higher 0.81 and our RMSE drops about 150 more. Not major, but it does help. Perhaps some better choices for the transformations would have more serious effect. In this case, I did not see any evidence that overfitting is an issue, at least for this dataset. It seemed that adding regressors for all the relevant attributes only made the accuracy of the model better. It is certainly possible we did not do enough to introduce overfitting issues, or perhaps the test training sets were not sufficient enough to show lack of accuracy in those cases.
